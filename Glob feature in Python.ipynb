```python
#Laoding the necessary libraries
import glob
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
def visualize(data):
    "The function creates the necessary charts to visualize the dataset"
    viz = pd.read_csv(data,header=None);#Reading the files
    a = viz.mean(axis=0);#mean value of features
    b = viz.max(axis=0);#Maximum of each features
    c = viz.min(axis=0);#Minimum value of features
    fig, ax = plt.subplots(figsize=(8,4)); #Create layout for plots
    ax.plot(b,label="Maximum");
    ax.plot(a,label="Mean");
    ax.plot(c,label="Minimum");
    fig.suptitle('Value vs Features Chart', fontsize=12)
    ax.set_xlabel('Feature Numbers', fontsize=10)
    ax.set_ylabel('Values', fontsize='medium')
    ax.legend();
    plt.show()
```

```python
def identify_issue(data):
    "The function tries to understand the issues in the datasets"
    ident = pd.read_csv(data,header=None); #Reading the dataset
    if (ident[0].max()==0 and ident[20].max()==20):
        print("I doubt the integirty of this dataset \n \n");
    elif (ident.min(axis=0).sum()==0):
        print("Why is this happening ? \n \n");
    else:
        print("Dataset looks good \n \n");
```

```python
#Reading the datasets from the folder
datasets = sorted(glob.glob('data*.csv'))
print(datasets)
```

```python
#Create the necessary plots and check the integrity of the dataset.
for check in datasets:
    print('\n \n \n',check)
    visualize(check);
    identify_issue(check);
```
